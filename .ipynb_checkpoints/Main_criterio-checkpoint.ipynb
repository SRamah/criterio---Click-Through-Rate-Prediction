```python
import pickle

def DumpObject(name, Object):
    # enregistrement ...
    with open(name, 'wb') as fichier:
        mon_pickler = pickle.Pickler(fichier)
        mon_pickler.dump(Object)

def LoadObject(name):
    # Lecture des objets contenus dans le fichier...
    with open(name, 'rb') as fichier:
        mon_depickler = pickle.Unpickler(fichier)
        object_recupere = mon_depickler.load()
    return object_recupere
```

```python
def generateFeatureEngine(DF , toGenerate):
    for g in toGenerate :
        for i in toGenerate :
            if i != g :
                newName = str(g) + ":" + str(i)
                DF[newName] = DF[i].astype('str')+":"+DF[g].astype('str')
                #DF[newName] = newName+DF[i].astype('str')+":"+DF[g].astype('str')
                #print newName
                for j in toGenerate :
                    if j!= i and j!=g :
                        newX = newName + ":" + str(j)
                        DF[newX] = DF[newName]+":"+DF[j].astype('str')
                        #DF[newX] = newX+DF[newName]+":"+DF[j].astype('str')
                        #print newX
```

```python
def getNameCols():
    num_Cols=[]
    cat_Cols = []
    dd = {}
    dd[0]="Label"
    for i in range(1,14):
        dd[i]="I"+`i`
        num_Cols.append("I"+`i`)
    for c in range(1,27):
        dd[c+13]="C"+`c`
        cat_Cols.append("C"+`c`)
    
    return dd, num_Cols, cat_Cols
```

```python
def PrepareData(DF, FutureEngineering):
    dd, num_Cols, cat_Cols = getNameCols()
    DF.rename(columns=dd, inplace=True)
    generateFeatureEngine(DF, FutureEngineering[:3])
    return np.array(DF, dtype='str'), DF.columns.values
```

```python
def logloss(p, y):
    p = max(min(p, 1. - 10e-17), 10e-17)        # The bounds
    return -log(p) if y == 1. else -log(1. - p)
```

```python
def get_direct_x(csv_row, header, D):
    ind = []
    x = {}
    for key in range(1,len(header)):
        if csv_row[key]!='nan':
            s = header[key] + '=' + str(csv_row[key])
        else:
            s = header[key] + '=' + ''
        ind.append(murmurhash3_32(s, positive=True) % D)
           
    x[0] = 1  # 0 is the index of the bias term
    for index in ind:
        if(not x.has_key(index)):
            x[index] = 0
        x[index] += 1
        
    return x 
```

```python
def get_test_x(csv_row, header, D):
    ind = []
    x = {}
    for key in range(0,len(header)):
        if csv_row[key]!='nan':
            s = header[key] + '=' + str(csv_row[key])
        else:
            s = header[key] + '=' + ''
        ind.append(murmurhash3_32(s, positive=True) % D)
           
    x[0] = 1  # 0 is the index of the bias term
    for index in ind:
        if(not x.has_key(index)):
            x[index] = 0
        x[index] += 1
        
    return x 
```

```python
def get_prob(x, w):
    wTx = 0.
    for i, xi in x.items():
        wTx += w[i] * xi  # w[i] * x[i]
        
    return 1. / (1. + exp(-max(min(wTx, 50.), -50.)))  # bounded sigmoid
```

```python
def update_w(w, g, x, p, y, alpha):
    for i, xi in x.items():
        g[i] += 1
        w[i] -= (p - y) * xi * alpha / (sqrt(g[i]))  # Minimising log loss
    return w, g
```

```python
def testing(df, FutureEngineering, D, w, t, submission):
    dd = {}
    for i in range(1,14):
        dd[i-1]="I"+`i`   
    for c in range(1,27):
        dd[c+12]="C"+`c`
    
    df.rename(columns=dd, inplace=True)
    generateFeatureEngine(df, FutureEngineering[:3])
    header = df.columns.values
    df = np.array(df, dtype='str')
    cx=0
    for row in df:
        
        # step 1, get the hashed features
        x = get_test_x(row, header, D)
        # step 2, get prediction
        p = get_prob(x, w)
        submission.write('%d,%f\n' % (60000000+int(t), p))
        t+=1
```

```python
def testAll(extractfile, chunksize, FutureEngineering, D, w):
    extractfile.seek(0)
    # build kaggle's submission file
    with open('submission.csv', 'w') as submission:
        submission.write('Id,Predicted\n')
        t=0
        for chunk in pd.read_csv(extractfile, chunksize=chunksize, header=-1, sep='\t'):
            #print "\ntesting ...",`(t+chunksize)/chunksize`,"M"
            testing(chunk, FutureEngineering, D, w, t, submission)
            t=t+len(chunk)
        print "\nDone \n"
            
        
```

```python
def training(ExFile, chunksize, stop_training, FutureEngineering, alpha, batch_size, D, LoadBack):
    
    # initialization
    weights = [0.] * D  # weights
    gradients = [.5] * D  # sum of historical gradients
    scores_liste = []
    loss=0
    lossb = 0.
    ter = 1
    
    # get data
    ExFile.seek(0)
    print "start training\n===============\n"
    stp=0
    for chunk in pd.read_csv(ExFile, chunksize=chunksize, header=-1, sep='\t'):
        
        if stp==stop_training:
            break
        stp=stp+1
        
        #print "\ntraining ...",(ter-1)/chunksize+1," x ",`chunksize`
    
        df_train, header = PrepareData(chunk, FutureEngineering)
        for row in df_train:
            y = 1. if row[0] == '1' else 0.
            # step 1, get the hashed features
            x = get_direct_x(row, header, D)
            # step 2, get prediction
            p = get_prob(x, weights)

            # for progress validation, useless for learning our model
            lossx = logloss(p, y)
            loss += lossx
            lossb += lossx
            if (ter % batch_size == 0 and ter > 1):
                print('\t Line : %d\t logloss: %f\t batch logloss: %f' % ( ter, loss/ter,  lossb/batch_size))
                lossb = 0.
                scores_liste.append(loss/ter)
                
            # step 3, update model with answer
            if ter>=1 and ter<1*(10**6):
                weights, gradients = update_w(weights, gradients, x, p, y, 0.13)
            elif ter>=1*(10**6) and ter<2*(10**6):
                weights, gradients = update_w(weights, gradients, x, p, y, 0.125)
            elif ter>=2*(10**6) and ter<3*(10**6):
                weights, gradients = update_w(weights, gradients, x, p, y, 0.12)
            elif ter>=3*(10**6) and ter<4*(10**6):
                weights, gradients = update_w(weights, gradients, x, p, y, 0.05)
            elif ter>=4*(10**6) and ter<=7*(10**6):
                weights, gradients = update_w(weights, gradients, x, p, y, 0.01)
            elif ter>7*(10**6) and ter<9*(10**6):
                weights, gradients = update_w(weights, gradients, x, p, y, 0.002)
            elif ter>=9*(10**6) and ter<21*(10**6):
                weights, gradients = update_w(weights, gradients, x, p, y, 0.0001)
            elif ter>=21*(10**6):
                weights, gradients = update_w(weights, gradients, x, p, y, 0.0001)
            
            #weights, gradients = update_w(weights, gradients, x, p, y, alpha)
            ter+=1
            
    #Save settings if LoadBack == True
    if LoadBack:
        DumpObject("settings/weights_"+`ter-1`, weights)
        DumpObject("settings/score_liste_"+`ter-1`, scores_liste)
    if (ter-1) % batch_size != 0:
        print('\t Line : %d\t  logloss: %f\t batch logloss: %f \n' % ( ter, loss/ter,  lossb/(ter%batch_size)))
        lossb = 0.
        scores_liste.append(loss/ter)

                       
    

    plt.plot([vx for vx in xrange(1,len(scores_liste)+1)], scores_liste)
    plt.title('Score Logloss')
    plt.xlabel('Training size * batch_size')
    plt.ylabel('Logloss')
    plt.grid(True)
                       
    return weights
```
